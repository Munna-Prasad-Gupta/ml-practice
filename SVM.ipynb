```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
import seaborn as sns

df = sns.load_dataset('iris')
df.head()
```

```python
df.columns

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['species_c'] = le.fit_transform(df['species'])

df.iloc[75]
```

```python
from matplotlib import pyplot as plt

# %matplotlib.inline

df0 = df[df['species_c'] == 0]
df1 = df[df['species_c'] == 1]
df2 = df[df['species_c'] == 2]

df0
```

```python
df0.columns
```

```python
import matplotlib.pyplot as plt

# Scatter plots for df0 and df1
plt.scatter(df0['sepal_length'], df0['sepal_width'], color='green', marker='+', label='Class 0')  # label for df0
plt.scatter(df1['sepal_length'], df1['sepal_width'], color='red', marker='*', label='Class 1')    # label for df1

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plots for df0 and df1
plt.scatter(df0['petal_length'], df0['petal_width'], color='green', marker='+', label='Class 0')  # label for df0
plt.scatter(df1['petal_length'], df1['petal_width'], color='red', marker='*', label='Class 1')    # label for df1

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.svm import SVC

clf = SVC(kernel ='linear')

X,y = df.iloc[:,:-2] ,df.iloc[:]['species_c']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.25)

clf.fit(X_train,y_train)

y_predicted = clf.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_predicted)
```
