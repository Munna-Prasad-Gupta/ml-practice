```python
import matplotlib.pyplot as plt

def step_activation(s):
  if s >= 0:
    return 1
  else:
    return 0

def sum_square_error(li):
  sum = 0
  for i in li:
    sum += i**2
  return sum

def call_nn(inp,out,w0,w1,w2,a):

    total_error = 1
    total_error_for_each_epoch = []
    for epoch in range(1000) :
      if total_error <= 0.02:
        break

      error_list = []
      predict_list =[]
      w0_correction = 0
      w1_correction = 0
      w2_correction = 0
      for idx,obs in enumerate(inp):
        Z = obs[0] * w1 + obs[1] * w2 + w0
        s = step_activation(Z)
        predict_list.append(s)
        error = s - out[idx]
        error_list.append(error)
        w0_correction += a * error
        w1_correction += a * error * obs[0]
        w2_correction += a * error * obs[1]

      total_error = sum_square_error(error_list)
      total_error_for_each_epoch.append(total_error)

      w0 -= w0_correction
      w1 -= w1_correction
      w2 -= w2_correction

    return w0,w1,w2,total_error_for_each_epoch,epoch



input = [[0,0],
         [0,1],
         [1,0],
         [1,1]]

output = [0,0,0,1]
W0 = 10
W1 = 0.2
W2=-0.75
alpha = 0.05

w0,w1,w2,total_error_for_each_epoch,epoch = call_nn(input,output,W0,W1,W2,alpha)

print(f"w0 = {w0} , w1 ={w1} , w2 = {w2} ,epoch = {epoch}")

plt.plot(total_error_for_each_epoch)
```
