```python
!pip install googletrans==3.1.0a0

import pandas as pd
from googletrans import Translator

# Load the Excel file containing Hindi sentences
data = pd.read_excel("split_part_4.xlsx")
print("Original Data:\n", data)

# Initialize the Translator
translator = Translator()

# Create a new DataFrame to store the original Hindi text and its English translation
translated_data = pd.DataFrame()

# Loop through each column and translate Hindi sentences to English
for column in data.columns:
    # Translate each row in the current column
    translated_column = []
    for element in data[column]:
        if isinstance(element, str) and element.strip():  # Only translate non-empty strings
            try:
                # Translate the Hindi text to English
                translated_text = translator.translate(element, src='hi', dest='en').text
                translated_column.append(translated_text)
            except Exception as e:
                print(f"Error translating '{element}': {e}")
                translated_column.append(element)  # In case of error, keep the original text
        else:
            translated_column.append(element)  # Keep non-string or empty values as they are

    # Add the original Hindi and translated English to the DataFrame
    translated_data[column + '_Hindi'] = data[column]  # Original Hindi text
    translated_data[column + '_English'] = translated_column  # Translated English text

# Save the translated data to a CSV file
translated_data.to_csv('spleng4.csv', index=False)

# Load the CSV file to inspect the first few rows
# data = pd.read_csv('english.csv')
# print("\nTranslated Data:\n", data.head())
```
